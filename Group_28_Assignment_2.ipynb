```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Housing.csv')
df.head()
```

```python
# Line Chart

sns.lineplot(x='stories', y='price', data=df, estimator='mean' )
plt.title('Average Price by Number of Stories')
plt.xlabel('No. of Stories')
plt.ylabel('Avg Price')
plt.title('Avg Price vs No. of Stories')
plt.show()
```

```python
# Scatter Plot

sns.scatterplot(x='area', y='price', data=df , hue = 'price' , palette= 'Greens_d'  )
plt.title('Average Price by Area')
plt.xlabel('Area' , fontsize = 15)
plt.ylabel('Avg Price' , fontsize = 15)
plt.show()
```

```python
# Scatter Plot

sns.barplot(x='furnishingstatus', y='price', data=df , estimator= 'mean')
plt.title('Furnishing Status by Price')
plt.xlabel('Furnishing Status' , fontsize = 15)
plt.ylabel('Price' , fontsize = 15)
plt.show()
```

```python
sns.histplot(data=df, x='price', bins=30)

# Customize the plot (optional)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
